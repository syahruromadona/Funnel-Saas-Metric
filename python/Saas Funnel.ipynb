```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\USER\Videos\Data Analysis Portfolio\Learn Data Cleaning\source_metric.csv")
df_dedup_table = pd.read_csv(r"C:\Users\USER\Videos\Data Analysis Portfolio\Learn Data Cleaning\dedup_table.csv")
```

```python
# Funnel efficiency
# Bar chart per source:
# trial_to_activation_rate (bar 1)
# activation_to_paid_rate (bar 2, side-by-side)
# Helps identify which source is efficient at moving users through the funnel

x = np.arange(len(df['source_std']))
width = 0.35

plt.bar(x-width/2, df['trial_to_activated_pct'],width, label='Trial to Activated')
plt.bar(x+width/2, df['activated_to_converted_pct'], width, label='Activated to Converted')

plt.xticks(x, df['source_std'])

plt.title('Funnel Efficiency by acquisition cost')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 2. **Activation speed**
# Box plot or horizontal bar:
# median_activation_days` per source

df_dedup_table.boxplot (
    column='activation_days',
    by='source_std',
    grid= False
)

plt.title('Activation Speed by Source')
plt.suptitle('')
plt.xlabel('')
plt.ylabel('Days')
plt.tight_layout()
plt.show()
```

```python
# **Revenue quality**
# Bar chart of `avg_mrr` per source

plt.bar(df['source_std'],df['avg_mrr'])
plt.title('Average MRR produced by sources')
plt.xlabel('Acquisition Source')
plt.ylabel('Average MRR (MYR)')
plt.tight_layout()
plt.show()
```
